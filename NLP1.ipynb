```python
import spacy
import pandas as pd
import gdown
import random
from faker import Faker
```

```python
fake = Faker()
nlp = spacy.load("en_core_web_sm")
```

```python
templates = [
    "The {} {} {} over the {} {}.",
    "{} is known for its {} and {} culture.",
    ...
]
```

```python
sentences = []

for _ in range(100):
    # Filter out non-string elements from templates before choosing
    valid_templates = [t for t in templates if isinstance(t, str)]
    if not valid_templates: # Handle case where all templates are invalid
        print("Warning: No valid string templates found. Skipping sentence generation.")
        break
    template = random.choice(valid_templates)
    words = [fake.word() for _ in range(template.count("{}"))]
    sentence = template.format(*words)
    sentences.append(sentence)

df = pd.DataFrame(sentences, columns=["text"])
```

```python
def process_texts(texts):
    pos_tags_list = []
    chunks_list = []

    for doc in nlp.pipe(texts, batch_size=20, disable=["ner"]):
        pos_tags = [(token.text, token.pos_, token.lemma_) for token in doc]
        chunks = [chunk.text for chunk in doc.noun_chunks]

        pos_tags_list.append(pos_tags)
        chunks_list.append(chunks)

    return pos_tags_list, chunks_list
```

```python
df["PoS_Tags_Lemmas"], df["Chunks"] = process_texts(df["text"])
```

```python
df.to_csv("input_sentences.csv", index=False)
print(df.head())
```
